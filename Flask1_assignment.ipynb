```python
# Q1.What is Flask Framework? What are the advantages of Flask Framework?
Ans.
Flask is a micro web framework written in Python. It is designed to be lightweight, flexible, and easy to use. 
Flask provides a set of tools and libraries that allow developers to build web applications quickly and efficiently.

The advantages of flask framework
1.Lightweight: Flask is a lightweight framework because it is independent of external libraries and it gives a quick 
start for web development having complex applications.
2.Compatible: Flask is compatible with the latest technologies such as machine learning, agile development, cloud 
technologies, etc.
3.Independent: Flask allows full control to the developers for creating web applications. A developer can do the experiment
with the libraries and architecture of the framework.
```

```python
# Q2.Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in
# Jupyter Notebook.
Ans.
from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
# Q3.What is App routing in Flask? Why do we use app routes?
Ans.
App Routing means mapping the URLs to a specific function that will handle the logic for that URL. Modern web frameworks
use more meaningful URLs to help users remember the URLs and make navigation simpler. 

When we create a Flask application, we can use the @app.route decorator to associate a URL pattern with a view function. 
The URL pattern is specified as a string parameter in the decorator, and the associated view function is defined immediately
below it. 
```

```python
# Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
#route to show the following details:
#Company Name: ABC Corporation
#Location: India
#Contact Detail: 999-999-9999
Ans.
from flask import Flask

app = Flask(__name__)

@app.route("/")
def company():
    return '''  
    <h1>Company Name: ABC Corporation</h1>
    <h2>Location: India</h2>
    <h2>Contact Detail: 999-999-9999</h2>
    '''

@app.route("/welcome")
def welcome():
    return "<h1>Welcome to the ABC Corporation</h1>"
    
if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
# Q5.What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the
# url_for() function.
Ans.
The url_for() function is used for URL building.

code:
from flask import Flask, redirect, url_for

app = Flask(__name__)

@app.route("/success/<score>")
def success(score):
    return "You have pass the exam by that marks {}".format(score)

@app.route("/fail/<score>")
def fail(score):
    return "You have failed the exam by that marks {}".format(score)

@app.route("/result/<int:marks>")
def result(marks):
    result = ""
    if marks < 50:
        result = 'fail'
    else:
        result = 'success'
    return redirect(url_for(result,score=marks))

if __name__ == "__main__":
    app.run(host="0.0.0.0")
```
